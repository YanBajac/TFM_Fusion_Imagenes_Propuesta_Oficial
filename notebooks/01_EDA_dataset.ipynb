# 01 – Exploración Inicial del Dataset (EDA)

Este notebook realiza un análisis exploratorio de los pares VIS/IR disponibles:
- Distribución de histogramas
- Estadísticas descriptivas
- Visualización de muestras representativas

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt
from src.datasets import list_pairs, load_pair
```

```python
pairs = list_pairs()
print(f'Total de pares encontrados: {len(pairs)}')
for vis_p, ir_p in pairs[:5]:
    print(f'  VIS: {vis_p.name}  |  IR: {ir_p.name}')
```

```python
# Visualizar el primer par
if pairs:
    vis, ir = load_pair(*pairs[0])
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    axes[0].imshow(vis, cmap='gray'); axes[0].set_title('VIS'); axes[0].axis('off')
    axes[1].imshow(ir,  cmap='gray'); axes[1].set_title('IR');  axes[1].axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# Histogramas comparativos
if pairs:
    fig, axes = plt.subplots(1, 2, figsize=(10, 3))
    axes[0].hist(vis.flatten(), bins=64, color='steelblue', edgecolor='none')
    axes[0].set_title('Histograma VIS')
    axes[1].hist(ir.flatten(),  bins=64, color='darkorange', edgecolor='none')
    axes[1].set_title('Histograma IR')
    plt.tight_layout()
    plt.show()
```
