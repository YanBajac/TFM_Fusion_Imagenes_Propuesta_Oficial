Carga los resultados de `experiments/results/metrics_reports/` y realiza:
- Estadísticos descriptivos por método
- Pruebas de hipótesis (Wilcoxon, Friedman)
- Gráficas comparativas (boxplot, barras)

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from src.utils import plot_metrics_bar

RESULTS_CSV = '../experiments/results/metrics_reports/all_metrics.csv'
```

```python
df = pd.read_csv(RESULTS_CSV)
print(df.head())
df.groupby('method')[['EN','SD','FE','MG','MI_vis','MI_ir']].describe().round(4)
```

```python
# Boxplot de Entropía por método
fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(data=df, x='method', y='EN', ax=ax)
ax.set_title('Entropía (EN) por método')
ax.set_xlabel('Método')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
# Prueba de Wilcoxon: Top-Hat vs. Promedio (EN)
th = df[df['method'].str.startswith('TopHat')]['EN'].values
av = df[df['method'] == 'Promedio']['EN'].values

if len(th) == len(av):
    stat, p = stats.wilcoxon(th, av)
    print(f'Wilcoxon TopHat vs Promedio (EN): stat={stat:.4f}, p={p:.4f}')
else:
    print('Grupos de distinto tamaño – revisa el CSV.')
```
